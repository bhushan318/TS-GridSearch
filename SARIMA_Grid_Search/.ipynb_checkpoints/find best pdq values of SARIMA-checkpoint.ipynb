```python
import sarima_grid as Sarima
import pandas as pd
```

```python
df = pd.read_csv('DailyDelhiClimateTest.csv')
```

```python
df['date'] = pd.to_datetime(df['date'])
df = df.set_index('date')
df.head()
```

```python
ts = df['meantemp']
n=10
train, test = ts[0:len(ts)-n], ts[len(ts)-n:]
```

```python
sarima_grid = Sarima.SARIMAGRID(train, test,seasonal_period = 7, max_value = 2)
```

```python
results =sarima_grid.sarimagridsearch()
results
```

- Training and testing MSE, MAE and MAPE are  evaluated along with model AIC and BIC
- Depending on time series sutiable matrices can be considered to find best trend and seasonality
- Default results are sorted based on AIC values. 

### Fit model optimal trend and seasonality

```python
import statsmodels.api as sm
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_absolute_percentage_error
import matplotlib.pyplot as plt

data = results[['p','d','q','P','D','Q','seasonal_period']]
data = data.iloc[0:1,]
p=int(data['p'])
d=int(data['d'])
q=int(data['q'])
P=int(data['P'])
D=int(data['D'])
Q=int(data['Q'])
s=int(data['seasonal_period'])
model_Sarima = sm.tsa.statespace.SARIMAX(endog=train, order=(p,d,q), 
                                            seasonal_order=(P,D,Q,s)).fit(disp=-1)
predictions_sarima = model_Sarima.predict(start=len(train), end=len(train)+len(test)-1, dynamic=False)
error_Sarima_mae_fitted = round(mean_absolute_error(train, model_Sarima.fittedvalues))
error_Sarima_mae_prediction = round(mean_absolute_error(test, predictions_sarima),2)
model_Sarima.summary()
```

```python
# training vs fitting plot
plt.figure(figsize=(15,7))
plt.plot(train, label="Test values")
plt.plot(train.index,model_Sarima.fittedvalues , label="Fitted values",color='red')
plt.title('SARIMA :Fitted MAE-%.3f' % error_Sarima_mae_fitted)
plt.legend(loc=1)
plt.show()
```

```python
# Actual vs prediciton 
plt.figure(figsize=(15,7))
plt.plot(test, label="Test values")
plt.plot(test.index, predictions_sarima, label="Predicted values",color='red')
plt.title('SARIMA: Testing MAE: %.3f' % error_Sarima_mae_prediction)
plt.legend(loc=1)
plt.show()
```
